```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 50
sd = 5
p1 = [5,3]
p2 = [-3,-4]
```

```python
x1 = np.random.normal(p1[0],sd,n)
y1 = np.random.normal(p1[1],sd,n)
x2 = np.random.normal(p2[0],sd,n)
y2 = np.random.normal(p2[1],sd,n)
df = []
for i in range(n):
    df.append([x1[i],y1[i],0])
    df.append([x2[i],y2[i],1])
col = ['r','b']
for i in range(n):
    plt.scatter(df[i][0],df[i][1],color = col[df[i][2]], s = 5)
plt.scatter(p1[0],p1[1],color = 'r', marker='+', s = 50)
plt.scatter(p2[0],p2[1],color = 'b', marker='+', s = 50)

plt.show
```

```python
# KNN function

def knn(x,y,k):
    #distance matrix:
    dist = []
    ind = []
    for i in range(n):
        cur_dist = (x-df[i][0])**2+(y-df[i][1])**2
        if i < k:
            dist.append(cur_dist)
            ind.append(i)
        else:
            cur_max = max(dist)
            if cur_dist < cur_max:
                ind[dist.index(cur_max)] = i
                dist[dist.index(cur_max)] = cur_dist
    return int(sum([df[x][2] for x in ind]) > (k-1)/2)
```

```python
# Plot the area
k = 5
for i in range(n):
    plt.scatter(df[i][0],df[i][1],color = col[df[i][2]])
plt.scatter(p1[0],p1[1],color = 'r', marker='x')
plt.scatter(p2[0],p2[1],color = 'b', marker='x')
for x in range(-15,15):
    for y in range(-15,15):
        plt.scatter(x, y, color = col[knn(x, y ,k)], s = 1)
plt.show
```
